```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1,2,3,4,5,6,7,8,])
plt.figure()
plt.subplot(1,2,1)
plt.plot(x)
plt.subplot(1,2,2)
plt.plot(x**2)
```

```python
x = np.array([1,2,3,4,5,6,7,8,])

plt.figure(figsize=(15,5))

ax1 = plt.subplot(1,3,1)
plt.plot(x)
ax2 = plt.subplot(1,3,2, sharey=ax1)
plt.plot(x**2)
ax3 = plt.subplot(1,3,3, sharey=ax1)
plt.plot(x**2//2)
```

# Histogramas

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2,2,sharex=True)
axs = [ax1,ax2,ax3,ax4]

for _ax in range(len(axs)):
    sample_size = 10**(_ax+1)
    sample = np.random.normal(loc=0.0, scale=1.0, size=sample_size)
    axs[_ax].hist(sample)
    axs[_ax].set_title(f'n={sample_size}')
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2,2,sharex=True)
axs = [ax1,ax2,ax3,ax4]

for _ax in range(len(axs)):
    sample_size = 10**(_ax+1)
    sample = np.random.normal(loc=0.0, scale=1.0, size=sample_size)
    axs[_ax].hist(sample, bins=100)
    axs[_ax].set_title(f'n={sample_size}')
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2,2,sharex=True)
axs = [ax1,ax2,ax3,ax4]

for _ax in range(len(axs)):
    sample_size = 10**(_ax+1)
    sample = np.random.normal(loc=0.0, scale=1.0, size=sample_size)
    axs[_ax].hist(sample, bins=5)
    axs[_ax].set_title(f'n={sample_size}')
```

```python
normal = np.random.normal(loc=0.0, scale=1.0, size=1000)
rand = np.random.random(size=1000) * 4 - 2

plt.figure()
plt.scatter(rand, range(len(normal)), c='red')
plt.scatter(normal, range(len(normal)))
```

```python
import matplotlib.gridspec as gridspec
plt.figure()
gspec = gridspec.GridSpec(3,3)

top = plt.subplot(gspec[0,1:])
side = plt.subplot(gspec[1:,0])
lower = plt.subplot(gspec[1,1:])
more_lower = plt.subplot(gspec[2:,1:])

lower.scatter(normal,rand)
top.hist(normal, bins=50)
side.hist(rand, bins=50, orientation='horizontal')

more_lower.scatter(rand, range(len(normal)), c='red')
more_lower.scatter(normal, range(len(normal)))
more_lower.clear()
more_lower.scatter(rand, range(len(normal)), c='red')
more_lower.set_xlim(-3,3)
more_lower.set_ylim(-100,1500)

side.invert_xaxis()
```

```python
import pandas as pd
import numpy as no
```

```python
normal_sample = np.random.normal(loc=0.0, scale=1.0, size=10000)
random_sample = np.random.random(size=10000)
gamma_sample = np.random.gamma(2, size=10000)
```

```python
df = pd.DataFrame({'normal': normal_sample,
             'random': random_sample,
             'gamma': gamma_sample,})
```

```python
df.describe()
```

```python
plt.figure()
plt.boxplot(df['normal'], whis='range')
```

```python
plt.figure(figsize=(16,10))
plt.boxplot([df['normal'], df['random'], df['gamma']])
```

```python
import mpl_toolkits.axes_grid1.inset_locator as mpl_il

plt.figure(figsize=(16,10))
plt.boxplot([df['normal'], df['random'], df['gamma']])
ax2 = mpl_il.inset_axes(plt.gca(), width='60%', height='40%', loc=2)
ax2.hist(df['gamma'], bins=100)
ax2.margins(x=0.2)
```
