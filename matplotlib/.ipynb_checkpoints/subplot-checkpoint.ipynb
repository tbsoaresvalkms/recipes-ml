```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1,2,3,4,5,6,7,8,])
plt.figure()
plt.subplot(1,2,1)
plt.plot(x)
plt.subplot(1,2,2)
plt.plot(x**2)
```

```python
x = np.array([1,2,3,4,5,6,7,8,])
plt.figure()
plt.subplot(1,2,1)
ax1 = plt.plot(x)
plt.subplot(1,2,2, sharey=ax1)
plt.plot(x**2)
```
